```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df_cs = pd.read_csv("Customers.csv")
df_ts = pd.read_csv("Transactions.csv")
```

```python
df_cs
```

```python
df_ts
```

```python
df = pd.merge(df_ts, df_cs, on='CustomerID')
```

```python
df_agg = df.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'TransactionID': 'count',
    'Region': 'first'
}).reset_index()
```

```python
df_ts
```

```python
df = df_cs.merge(df_ts[['CustomerID', 'Total_Spend', 'Average_Spend', 'Recency', 'Frequency']], on='CustomerID', how='left')
df
```

```python
df_agg = df.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'TransactionID': 'count',
    'Region': 'first'
}).reset_index()
```

```python
df_agg = pd.get_dummies(df_agg, columns=['Region'], drop_first=True)
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_agg.drop(columns=['CustomerID']))
```

```python
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(df_scaled)
```

```python
df_agg['Cluster'] = clusters
```

```python
from sklearn.metrics import davies_bouldin_score, silhouette_score
db_index = davies_bouldin_score(df_scaled, clusters)
silhouette_avg = silhouette_score(df_scaled, clusters)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalValue', y='Quantity', hue='Cluster', data=df_agg, palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Total Value')
plt.ylabel('Quantity')
plt.legend(title='Cluster')
plt.show()
```

```python
print(f"Number of clusters formed: {kmeans.n_clusters}")
print(f"Davies-Bouldin Index: {db_index:.4f}")
print(f"Silhouette Score: {silhouette_avg:.4f}")
```
