```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.merge(df_ts, df_cs, on='CustomerID')
df = pd.merge(df, df_pd, on='ProductID')
df
```

```python
df.drop(['Price_x'], axis=1)
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['SignupDate'] = pd.to_datetime(df['SignupDate'])
df['CustomerAge'] = (df['TransactionDate'] - df['SignupDate']).dt.days
```

```python
customer_df = df.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'CustomerAge': 'max',
    'Price_y': 'mean',
    'ProductID': lambda x: x.nunique()
}).reset_index()
```

```python
scaler = StandardScaler()
customer_df[['TotalValue', 'Quantity', 'Price_y', 'ProductID']] = scaler.fit_transform(
    customer_df[['TotalValue', 'Quantity', 'Price_y', 'ProductID']])
```

```python
similarity_matrix = cosine_similarity(customer_df.iloc[:, 1:])
similarity_df = pd.DataFrame(similarity_matrix, index=customer_df['CustomerID'], columns=customer_df['CustomerID'])
```

```python
customer_ids = similarity_df.index[:20]
lookalikes = {}
for customer_id in customer_ids:
    similar_customers = similarity_df.loc[customer_id].nlargest(4)[1:4]  # Exclude self
    lookalikes[customer_id] = [(cust_id, round(similarity_df.loc[customer_id, cust_id], 2)) for cust_id in similar_customers.index]
```

```python
with open('NMK_Indudhar_Lookalike.csv', 'w') as file:
    for key, value in lookalikes.items():
        file.write(f'{key},{value}\n')
```
