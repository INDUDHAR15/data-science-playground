# Import Necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor


import seaborn as sns 
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from sklearn.naive_bayes import GaussianNB

#from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, r2_score
```

# Loading Dataset 

```python
df = pd.read_csv('migration_nz.csv')
df
```

# Identifying Unique Values of Measure

```python
df['Measure'].unique()
```

Replacing each unique string value its unique integer value

```python
df['Measure'].replace("Arrivals",0,inplace=True)
df['Measure'].replace("Departures",1,inplace=True)
df['Measure'].replace("Net",2,inplace=True)
```

```python
df['Measure'].unique()
```

```python
df['Country'].unique()
```

```python
df['CountryID'] = pd.factorize(df.Country)[0]
df['CitID'] = pd.factorize(df.Citizenship)[0]
```

```python
df['CountryID'].unique()
```

Identifying Null Values

```python
df.isnull().sum()
```

Handling by null values by fill with median value

```python
df["Value"].fillna(df["Value"].median(),inplace=True)
```

```python
df.isnull().sum()
```

# Split The Data into Train and Test sets

 I will split the data into 70 % training and 30 % test set

```python
X= df[['CountryID','Measure','Year','CitID']].to_numpy()
Y= df['Value'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(
  X, Y, test_size=0.3, random_state=9)
```

# Predict Migration

By using Linear Regression 

```python
lr_model = LinearRegression()

# Train the model
lr_model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = lr_model.predict(X_test)

# Calculate the mean squared error and R^2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_pred, y_test)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
print(f'Linear Regression Train Score: {lr_model.score(X_train,y_train)}')
print(f'Linear Regression Test Score: {lr_model.score(X_test,y_test)}')
```

By using Random Forest Regressor 

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)

# Train the model
rf_model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_rf = rf_model.predict(X_test)

# Calculate metrics
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f'Random Forest - Mean Squared Error: {mse_rf}')
print(f'Random Forest - RÂ² Score: {r2_rf}')
print(f'Random Forest Regression Train Score: {rf_model.score(X_train,y_train)}')
print(f'Random Forest Regression Test Score: {rf_model.score(X_test,y_test)}')
```
